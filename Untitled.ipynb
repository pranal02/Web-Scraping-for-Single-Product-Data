```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL = "https://www.amazon.in/s?k=playstation+4&crid=17N5JN6BNTWLZ&sprefix=playstation+%2Caps%2C259&ref=nb_sb_noss_2"
```

```python
HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36','Accept-Language':'en-US,en;q=0.5'})
```

```python
webpage= requests.get(URL,headers=HEADERS)
```

```python
type(webpage.content)
```

```python
soup = BeautifulSoup(webpage.content,"html.parser")
```

```python
soup
# information in the form of html page will appear 
```

```python
links = soup.find_all("a",attrs={'class':'a-link-normal s-no-outline'})
```

```python
links
#here we get all the links associate with a tag ie link tag with class name a-link-normal s-no-outline
```

```python
#get the first link then extract the content and loop the same processor for the remaining links
link  =links[0].get('href')
```

```python
product_list = "https://amazon.in"+link
```

```python
product_list
#link of details of 1st product 
```

```python
new_webpage= requests.get(product_list,headers=HEADERS)
```

```python
new_webpage
```

```python
new_soup = BeautifulSoup(new_webpage.content,"html.parser")
```

```python
new_soup
#here we are able to extract the 1st product details in the html format
```

```python
new_soup.find("span",attrs={"id":'productTitle'}).text.strip()
# product title
```

```python
new_soup.find("span",attrs={"class":'a-price-whole'}).text
#product prince
```

```python
new_soup.find("span",attrs={"class":'a-icon-alt'}).text
#product rating
```
